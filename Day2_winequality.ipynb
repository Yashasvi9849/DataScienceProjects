```python
column_name = 'volatile acidity'
null_count = df[column_name].isnull().sum()
print(f"Number of null values in column '{column_name}': {null_count}")
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        df[col]= df[col].fillna(df[col].mean())
```

```python
df.quality.unique()
```

```python
df['best quality'] = [ 1 if x>5 else 0 for x in df['quality']]
```

```python
df.head()
```

```python
df.drop(columns = 'quality')
```

```python
#label encoder for categorical problem
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['type']=le.fit_transform(df['type'])
```

```python
#training and test 
y=df['best quality']
x = df.drop(columns='best quality')
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size = 0.2,random_state = 89)
x_train.shape,x_test.shape
```

```python
#Normalize data
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler()
x_train = norm.fit_transform(x_train)
from sklearn.linear_model import LogisticRegression 

lr_model = LogisticRegression()
lr_model.fit(x_train,y)
x_test = norm.fit_transform(x_train)
lr_prediction= lr_model.predict(x_test)
```

```python


from sklearn.metrics import accuracy_score
lr_prediction

lr_accuracy = accuracy_score(y_test,lr_prediction)
lr_accuracy
```
