```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sample = np.array([3,4,5,1,2,3,4,5,6,3,3,3,5,6,7,1,1,1,3,3,3,5])
sns.histplot(sample)
```

```python
#plotting the sample with kde
#bw is the bandwidth , if you increase the bw, the curve will become more smooth.
sample = np.array([3,4,5,1,2,3,4,5,6,3,3,3,5,6,7,1,1,1,3,3,3,5])
sns.kdeplot(sample,bw=0.1)
```

```python
car = sns.load_dataset('mpg')
car.head()
```

```python
car.cylinders.nunique()
```

```python
car.cylinders.unique()
```

```python
sns.histplot(car.cylinders)
```

```python
sns.kdeplot(car.cylinders , shade = False)
plt.title('KDE of Car Cylinders')
```

```python
car.isnull().sum()
```

```python
car = car.dropna()
```

```python
sns.kdeplot(car.horsepower)
```

```python
sns.kdeplot(car.horsepower, bw =0.002)
```

```python
#plotting the 2 dimensional data
sns.kdeplot(data = car , x = car.mpg,y=car.horsepower)
```

```python
#plotting the horse power and mpg based on the number of cylinders present
sns.kdeplot(data = car , x=car.mpg,y=car.horsepower,hue = car.cylinders)
```

```python
plt.figure(figsize=(20,15))
sns.kdeplot(data = car , x = car.mpg,y=car.horsepower,hue = car.cylinders,shade = True)
```
