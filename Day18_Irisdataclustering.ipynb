importing the visualization libraries

```python
#importing the visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#accessing the Iris dataset from the seaborn library
iris = sns.load_dataset('iris')
iris.head()
```

```python
#visualizing the data points using different features in scatter plot
sns.scatterplot(iris,x='sepal_length',y='sepal_width',hue = 'species')
```

```python
sns.scatterplot(iris,x='petal_length',y='petal_width',hue = 'species')
```

```python
sns.scatterplot(iris,x='sepal_length',y='petal_length',hue = 'species')
```

```python
#hue='species' tells Seaborn to use the 'species' column to color-code the data points based on the different species of iris flowers.
sns.scatterplot(iris,x='sepal_width',y='petal_width',hue = 'species')
```

```python
#here we are splitting the clusters into 3,hence the values 0,1,2. in the output.each data point is assigned to a different set of cluster.
X = iris[['sepal_length','petal_length']].values
from sklearn.cluster import KMeans,DBSCAN
km = KMeans(n_clusters = 3)
km.fit(X)
km.predict(X)
```

```python
#km.predict() will give the clusters made by our Kmeans model.Now insted of giving the hue parameter the species column from the iris dataset we are going to give it the km.predict
#he scatter plot will color-code the data points based on their assigned cluster labels from the KMeans model, rather than their actual species labels.
sns.scatterplot(iris,x='sepal_length',y='petal_length',hue = km.predict(X))
```

```python
#dbscan
#dbscan has clustered our features into two clusters
dbscan = DBSCAN(eps =1,min_samples = 6)
dbscan.fit(X)

dbscan.labels_
```

```python
#visualizing the dbscan
sns.scatterplot(iris,x='sepal_length',y='petal_length',hue = dbscan.labels_)
```

```python
#PCA
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml('mnist_784',version = 1)
#scale pixel values to between 0 and 1
X=mnist.data/255.0
y=mnist.target

#reduce dimensionality using pca
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca

```

```python
#plotting the graph
plt.figure(figsize = (10,8))
plt.scatter(X_pca[:,0] , X_pca[:,1] ,c=y.astype(int),cmap = 'tab10',alpha = 0.5)
plt.colorbar(label = 'Digit Label')
plt.xlabel('Principal component 1 ')
plt.ylabel('Principal component 2')
plt.title('MNIST Dataset Scatter Plot (PCA)')
plt.grid(True)
plt.show()
```
