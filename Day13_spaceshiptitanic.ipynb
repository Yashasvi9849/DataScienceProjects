```python
import pandas as pd 
train = pd.read_csv('spaceship_train.csv')
test=pd.read_csv('spaceship_test.csv')
```

```python
train.shape,test.shape
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
#Handling null values
train.drop(columns=['Name','PassengerId'],inplace = True)
test.drop(columns=['Name','PassengerId'],inplace = True)
```

```python
train.head(30)
```

```python
train.tail(30)
```

```python
#replacing null value with its previous value with ffill() method .
train[['HomePlanet','CryoSleep','Destination','VIP']]=train[['HomePlanet','CryoSleep','Destination','VIP']].fillna(method = 'ffill')
test[['HomePlanet','CryoSleep','Destination','VIP']]=train[['HomePlanet','CryoSleep','Destination','VIP']].fillna(method = 'ffill')
```

```python
train.isnull().sum()
```

```python
#splitting the cabin into 3 columns cabindeck,cabinnum,cabinsize
train[['CabinDeck' , 'CabinNum' , 'CabinSize']]=train['Cabin'].str.split('/',expand = True)
test[['CabinDeck' , 'CabinNum' , 'CabinSize']]=train['Cabin'].str.split('/',expand = True)
```

```python
#checking the unique values in the dataset
train.nunique()
```

```python
#dropping the columns cabin,cabinNum
train.drop(columns =['Cabin','CabinNum'],inplace =True)
test.drop(columns =['Cabin','CabinNum'],inplace =True)
```

```python
#the cabinsize and cabindeck columns tend to repeat their previous values,use ffill
train['CabinDeck']=train['CabinDeck'].fillna(method='ffill')
train['CabinSize']=train['CabinSize'].fillna(method='ffill')

test['CabinDeck']=test['CabinDeck'].fillna(method='ffill')
test['CabinSize']=train['CabinSize'].fillna(method='ffill')
```

```python
#histogram only plots numerical values
train.hist(figsize=(20,10))
```

```python
#only age column has values spreading across the wide range,so we are imputing it with mean
train['Age']=train['Age'].fillna(train['Age'].mean())
test['Age']=test['Age'].fillna(test['Age'].mean())
```

```python
#filling all other numerical columns with mode
train['RoomService']=train['RoomService'].fillna(train['RoomService'].mode()[0])
test['RoomService']=test['RoomService'].fillna(test['RoomService'].mode()[0])

train['FoodCourt']=train['FoodCourt'].fillna(train['FoodCourt'].mode()[0])
test['FoodCourt']=test['FoodCourt'].fillna(test['FoodCourt'].mode()[0])

train['ShoppingMall']=train['ShoppingMall'].fillna(train['ShoppingMall'].mode()[0])
test['ShoppingMall']=test['ShoppingMall'].fillna(test['ShoppingMall'].mode()[0])

train['Spa']=train['Spa'].fillna(train['Spa'].mode()[0])
test['Spa']=test['Spa'].fillna(test['Spa'].mode()[0])

train['VRDeck']=train['VRDeck'].fillna(train['VRDeck'].mode()[0])
test['VRDeck']=test['VRDeck'].fillna(test['VRDeck'].mode()[0])
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train.dtypes
```

```python
#converting the cyrosleep , VIP,Transported to true  to 1 or false to 0
train[['CryoSleep','VIP','Transported']] = train[['CryoSleep','VIP','Transported']].astype(int)
test[['CryoSleep','VIP']] = test[['CryoSleep','VIP']].astype(int)
```

```python
#converting the categorical column into numerical column by using onehotencoding using get_dummies() method
df=pd.get_dummies(train,columns=['HomePlanet','Destination','CabinDeck','CabinSize'])
pd.set_option('display.max_columns',100)
df
```

```python
#when we are using classifier we donot need for any standardization.
y=df['Transported'].copy()
X=df.drop(columns='Transported').copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 786)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
dtc_prediction = dtc.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix
dtc_accuracy = accuracy_score(y_test,dtc_prediction)
dtc_accuracy
```

```python
#confusion Matrix 
dtc_confusion_matrix = confusion_matrix(y_test,dtc_prediction)
dtc_confusion_matrix
```

```python
#Random Forest Classifier 
from sklearn.ensemble import RandomForestClassifier 
rfc=RandomForestClassifier(random_state = 786)
rfc.fit(X_train,y_train)
rfc_prediction = rfc.predict(X_test)

rfc_accuracy = accuracy_score(y_test,rfc_prediction)
rfc_accuracy
```

```python
#confusion matrix
rfc_confusion_matrix = confusion_matrix(y_test,rfc_prediction)
rfc_confusion_matrix
```

```python
#XGB classifier 
from xgboost import XGBClassifier
xgbc=XGBClassifier(random_state = 786)
xgbc.fit(X_train,y_train)
xgbc_prediction=xgbc.predict(X_test)

xgbc_accuracy = accuracy_score(y_test,xgbc_prediction)
xgbc_accuracy
```
