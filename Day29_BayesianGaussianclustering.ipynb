```python
#clustering using baysesian gaussian mixture
import seaborn as sns
iris = sns.load_dataset('iris')
X=iris.drop(columns = 'species')
y = iris['species']
```

```python
sns.scatterplot(iris,x='petal_length',y='sepal_length',hue = 'species')
```

```python
from sklearn.mixture import BayesianGaussianMixture
bggm = BayesianGaussianMixture(n_components=3,random_state = 42)
bggm.fit(X)
y_pred = bggm.predict(X)
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt
plt.scatter(X['petal_length'],X['sepal_width'], c=y_pred,cmap = 'viridis')
```
