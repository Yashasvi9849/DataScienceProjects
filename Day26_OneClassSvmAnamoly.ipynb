```python
#Anamoly Detection using oneclassSVM 
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
#considering only columns sepal with and petal width 
sns.scatterplot(df, x=df.sepal_width,y=df.petal_width)
```

```python
#import one class svm 

from sklearn.svm import OneClassSVM 
ocsvm = OneClassSVM(nu = 0.01)# if you increase the nu value you will find more outliers ,decrease less outliers 
X= df[['sepal_width','petal_width']]
ocsvm.fit(X)
ocsvm.predict(X)
```

```python
sns.scatterplot(df,x=df.sepal_width, y = df.petal_width  , hue=ocsvm.predict(X), palette = 'Set1')
```
